```python
import seaborn as sns
```

```python
df = sns.load_dataset("titanic")
```

```python
df
```

```python
### checking the null values:

df.isnull().sum()
```

```python
### delete the rows having the missing values:

df.dropna()
```

```python
### columnwise dropping of null values:

df.dropna(axis = 1)
```

```python
### imputation of missing values:
```

```python
### mean value imputation:
```

```python
sns.distplot(df["age"])
```

```python
### since tje distribution of the values is nearly normal one, we are going to impute
### the missing values as the mean of the values of entire column:

df["age_mean"] = df["age"].fillna(df["age"].mean())
```

```python
df[["age", "age_mean"]]
```

```python
### note: mean imputation works well when we have normally distributed data.
```

```python
### median value imputation in case of presence of outliers
# or
# in case of left/ right skewed data:
```

```python
df["age_median"] = df["age"].fillna(df["age"].median())
```

```python
df[["age", "age_median", "age_mean"]]
```

```python
### mode imputation techniques for categorical features:
```

```python
df[df["embarked"].isnull()]
```

```python
df["embarked"].unique()
```

```python
df["embarked"].notna()
```

```python
df["embarked"].value_counts()
```

```python
df["embarked_mode"] = df["embarked"].fillna(df["embarked"].mode())
```

```python
df[["embarked", "embarked_mode"]]
```
